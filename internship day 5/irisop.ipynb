```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("iris (1).csv")

# First 5 rows
print(df.head())

# Last 5 rows
print(df.tail())

# Shape
print(df.shape)

# Column names
print(df.columns)

# Dataset info
print(df.info())

# Statistical summary
print(df.describe())

# Check null values
print(df.isnull().sum())

# Check duplicates
print(df.duplicated().sum())

# Remove duplicates
df.drop_duplicates(inplace=True)

# Fill missing values
df.fillna(0, inplace=True)

# Data types
print(df.dtypes)

# Unique values
print(df["species"].unique())

# Value counts
print(df["species"].value_counts())

# Sorting
print(df.sort_values(by="sepal_length"))

# Filtering
print(df[df["sepal_length"] > 5])

# Multiple conditions
print(df[(df["sepal_length"] > 5) & (df["species"] == "setosa")])

# Groupby
print(df.groupby("species").mean())

# Correlation
numeric_df = df.select_dtypes(include=np.number)
print(numeric_df.corr())

# LINE PLOT
plt.plot(df["sepal_length"])
plt.title("Line Plot")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.show()

# HISTOGRAM
plt.hist(df["petal_length"], bins=20)
plt.title("Histogram")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()

# BAR GRAPH
df["species"].value_counts().plot(kind='bar')
plt.title("Bar Graph")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()

# SCATTER PLOT
plt.scatter(df["sepal_length"], df["petal_length"])
plt.title("Scatter Plot")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()

# PIE CHART
df["species"].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Pie Chart")
plt.ylabel("")
plt.show()

# BOXPLOT
plt.boxplot(df["sepal_length"])
plt.title("Boxplot")
plt.show()

# HEATMAP
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap")
plt.show()

# COUNTPLOT
sns.countplot(x="species", data=df)
plt.title("Countplot")
plt.show()

# BOXPLOT
sns.boxplot(x="species", y="sepal_length", data=df)
plt.title("Seaborn Boxplot")
plt.show()

# VIOLIN PLOT
sns.violinplot(x="species", y="petal_length", data=df)
plt.title("Violin Plot")
plt.show()

# PAIRPLOT
sns.pairplot(df, hue="species")
plt.show()

# HISTPLOT
sns.histplot(df["petal_width"], kde=True)
plt.title("Histplot")
plt.show()

# KDE PLOT
sns.kdeplot(df["petal_length"], fill=True)
plt.title("KDE Plot")
plt.show()

# SCATTERPLOT
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=df)
plt.title("Seaborn Scatterplot")
plt.show()

# REGPLOT
sns.regplot(x="sepal_length", y="petal_length", data=df)
plt.title("Regression Plot")
plt.show()

df.to_csv("cleaned_iris.csv", index=False)

print("Analysis Completed")
```
